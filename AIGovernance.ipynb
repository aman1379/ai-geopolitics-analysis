```python
import pandas as pd
import random

# Create fake dataset
data = []

for i in range(100):
    actual = random.choice([0, 1])  # 0 = civilian, 1 = threat

    # AI prediction with some error
    if random.random() > 0.2:  # 80% accuracy
        predicted = actual
    else:
        predicted = 1 - actual  # wrong prediction

    data.append([actual, predicted])

df = pd.DataFrame(data, columns=["Actual", "Predicted"])

# Calculate errors
false_positives = len(df[(df["Actual"] == 0) & (df["Predicted"] == 1)])
false_negatives = len(df[(df["Actual"] == 1) & (df["Predicted"] == 0)])

print("False Positives (Civilian flagged as threat):", false_positives)
print("False Negatives (Threat missed):", false_negatives)
```
